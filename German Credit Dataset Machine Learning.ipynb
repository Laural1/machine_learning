When a bank receives a loan application, based on the applicant’s profile the bank has to make a decision regarding whether to go ahead with the loan approval or not. Two types of risks are associated with the bank’s decision –

If the applicant is a good credit risk, i.e. is likely to repay the loan, then not approving the loan to the person results in a loss of business to the bank
If the applicant is a bad credit risk, i.e. is not likely to repay the loan, then approving the loan to the person results in a financial loss to the bank
Objective of Analysis:

Minimization of risk and maximization of profit on behalf of the bank.

To minimize loss from the bank’s perspective, the bank needs a decision rule regarding who to give approval of the loan and who not to. An applicant’s demographic and socio-economic profiles are considered by loan managers before a decision is taken regarding his/her loan application.

The German Credit Data contains data on 20 variables and the classification whether an applicant is considered a Good or a Bad credit risk for 1000 loan applicants.  

A predictive model developed on this data is expected to provide a bank manager guidance for making a decision whether to approve a loan to a prospective applicant based on his/her profiles.

#### Convert Purpose feature from categorial to numerical. In addition, the values will be grouped by below iteration. This will help improve prediction accuracy later.

```python
def set_label(pur):
    purpose = 0
    home_related = ['A42','A44','A45',]
    new_car = ['A40']
    used_car = ['A41']
    other = ['A43','A46','A47','A48','A49','A410',]
    if pur in home_related:
        purpose = 1
    elif pur in new_car:
        purpose = 2
    elif pur in used_car:
        purpose = 3
    else:
        purpose = 4
    return purpose
```

```python
df['purpose_label']= df['purpose'].apply(lambda x: set_label(x))
```

```python
df['purpose_label']
```

### Perform the same PCA process to the account status feature: reduce the number of values by grouping them into: no account, have account with zero balance, have account with some balance. This will help improve Machine learning models' accuracry later. 

```python
def set_label(acc):
    account = 0
    zero_balance = ['A11']
    no_account = ['A14']
    other = ['A12','A13']
    if acc in zero_balance:
        account = 1
    elif acc in no_account:
        account = 2
    else:
        account = 3
    return account
df['Acct_Status_label']= df['Acct_Status'].apply(lambda x: set_label(x))
```

### Similar for credit history feature: group the values into 4 bins, and convert the column from categorial to numerical. 

```python
def set_label(cre):
    credit = 0
    no_credit = ['A30']
    paid_up = ['A31','A32']
    some_prob = ['A33']
    other = ['A34']
    if cre in no_credit:
        credit = 1
    elif cre in paid_up:
        credit = 2
    elif cre in some_prob:
        credit = 3
    else:
        credit = 4
    return credit
df['credit_his_label']= df['credit_his'].apply(lambda x: set_label(x))
```

### PCA - saving acccount: group the values into no saving, small saving, medium saving, and others. 

```python
def set_label(sav):
    saving = 0
    no_saving = ['A65']
    small_saving = ['A61']
    med_saving = ['A62','A63']
    other = ['A64']
    if sav in no_saving:
        saving = 1
    elif sav in small_saving:
        saving = 2
    elif sav in med_saving:
        saving = 3
    else:
        saving = 4
    return saving
df['saving_acct_label']= df['saving_acct'].apply(lambda x: set_label(x))
```

# K-Means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
mms = MinMaxScaler()
mms.fit(df[['sex_marital_trans','creditability']])
data_transformed = mms.transform(df[['sex_marital_trans','creditability']])
```

## use elbow method to determine k 

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
X1=df[['sex_marital_trans','creditability']].values
kmeans = KMeans(n_clusters=3, random_state=0).fit(X1)
y_kmeans = kmeans.predict(X1)
plt.scatter(X1[:, 0], X1[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```
